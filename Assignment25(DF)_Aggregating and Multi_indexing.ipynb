```python
import numpy as np
import pandas as pd
path = 'C:/Users/Abdullah Rashid/Desktop/Data Analysis with Python Udemy Course/Pandas+Course+Resources/Pandas Course Resources/retail/transactions.csv'
trans = pd.read_csv(path)
trans
```

```python
trans.groupby('store_nbr')[['transactions']].sum().sort_values('transactions',ascending = False).head(10)
```

```python
trans['month'] = trans['date'].astype('datetime64').dt.month
trans
```

```python
trans.groupby(['store_nbr','month'],as_index = False)[['transactions']].sum().sort_values(['month','transactions'],ascending=[True,False])
```

```python
grouped = (trans.groupby(['store_nbr','month'])
           .agg({'transactions' : ['sum','mean']})
           .sort_values(by=['month',('transactions','sum')], ascending = [True, False]))
grouped
```
