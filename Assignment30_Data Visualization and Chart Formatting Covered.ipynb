```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style("darkgrid")
path = 'C:/Users/Abdullah Rashid\Desktop/Data Analysis with Python Udemy Course/Pandas+Course+Resources/Pandas Course Resources/retail/transactions.csv'
trans = pd.read_csv(path)
trans_store1234 = trans.loc[trans['store_nbr'].isin([1,2,3,4])]
trans_store1234.pivot_table(index='date',columns = 'store_nbr',values = 'transactions').plot(
    subplots = True,
    layout=(2,2),
    sharey = True,
    figsize=(8,8),
    grid=True,
    xlabel = 'date',
    ylabel = 'transactions',
    title = ['Store1_transactions','Store2_transactions','Store3_transactions','Store4_transactions'],
    rot = 45)
```

```python
trans_store1234 = trans.loc[trans['store_nbr'].isin([1,2,3,4])]
trans1_store1234 = trans_store1234.pivot_table(index = 'date',
                                               columns = 'store_nbr',
                                               values = 'transactions')
trans1_store1234.index = trans1_store1234.index.astype('datetime64[ns]')
trans1_store1234
```

```python
trans1_store1234_by_month = trans1_store1234.groupby(trans1_store1234.index.month).sum()
trans1_store1234_by_month.plot(kind='barh',
                               figsize = (8,8),
                               rot = 0,
                               stacked = True,
                               xlabel = 'Months',
                               ).legend(loc = 'best')
```
