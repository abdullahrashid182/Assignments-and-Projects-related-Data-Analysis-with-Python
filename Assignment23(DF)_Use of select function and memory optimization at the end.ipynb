```python
import numpy as np
import pandas as pd
path = 'C:/Users/Abdullah Rashid/Desktop/Data Analysis with Python Udemy Course/Pandas+Course+Resources/Pandas Course Resources/retail/transactions.csv'
trans = pd.read_csv(path)
trans['pct_to_target'] = trans['transactions']/2500 # arithmetic column is added
trans['met_target'] = trans['pct_to_target'] >= 1 # boolean column is added
trans['bonus_payable'] = trans['met_target']*100 # arithmatic column is added
trans['month'] = trans['date'].astype('datetime64').dt.month # month column is added
trans['day of month'] = trans['date'].astype('datetime64').dt.day # day column is added
trans['day of week'] = trans['date'].astype('datetime64').dt.dayofweek
trans
```

```python
conditions = [(trans['month']==12),((trans['day of week']==6) & (trans['month']==5)), ((trans['day of week']==0) & (trans['month']==7))]
choices = ['Holiday Bonus','Corporate Month','Summer Special']
trans['Seasonal Bonus'] = np.select(conditions,choices,default='None')
trans['Seasonal Bonus'].value_counts() # trans.loc[:,'Seasonal Bonus'].value_counts() will give same result as well
```
